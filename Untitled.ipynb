```python


signals = {
    "buy_signals": [
        {
            "name": "bsig1",
            "form": ("sma1", lambda x, y: x>y, "sma2")
        },
        {
            "name":"bsig2",
            "form": ("rsi", lambda x, y: x>y, 30)
        }
        
    ],
    "sell_signals": [
    
        {
            "name": "ssig1",
            "form": ("sma1", lambda x, y: x<y, "sma2")
        },
        {
            "name":"ssig2",
            "form": ("rsi", lambda x,y : x<y, 70)
        }
    ]
}

indicators = [
        {
            "type": "sma",
            "name": "sma1",
            "args": (20,)
        },
        {
            "type": "sma",
            "name": "sma2",
            "args": (40,)
        },
        {
            "name": "sma3",
            "type": "sma",
            "args": (80,)
        },
        {
            "name": "rsi",
            "type": "rsi",
            "args": ()
        }]

def process_indicators(df, indicators):
    # indicators
    for ind in indicators:
        name = ind["name"]
        type_ = ind["type"]
        args = ind["args"]
        df[name] = getattr(df.ta, type_)(*args)
    return df

def process_signals(df, signals):
    # buy signals
    for bsig in signals["buy_signals"]:
        name = bsig["name"]
        form = bsig["form"]
        if type(form[2]) == str:
            df[name] = form[1](df[form[0]], df[form[2]])
            
        elif type(form[2]) == int:
            df[name] = form[1](df[form[0]], form[2])
            
    # buy signals
    for ssig in signals["sell_signals"]:
        name = ssig["name"]
        form = ssig["form"]
        if type(form[2]) == str:
            df[name] = form[1](df[form[0]], df[form[2]])
            
        elif type(form[2]) == int:
            df[name] = form[1](df[form[0]], form[2])
    return df

def plot_strategy(df, w, b):
    usdt_0 = usdt = 100
    btc = 0
    bought=False
    n_trades = 0
    
    w_b = w[:2]
    w_s = w[2:]
    b_b = b[:2]
    b_s = b[2:]
    
    buy_and_hold_pct =  (df.close[-1] - df.close[0]) / df.close[0]
    buy_time = []
    buy_price = []
    sell_time = []
    sell_price = []
    profit = []
        
    for index, row in df.iterrows():
        x_b = np.array([row["bsig1"], row["bsig2"]])
        x_s = np.array([row["ssig1"], row["ssig2"]])

        h_b = np.sum(w_b * x_b + b_b)
        h_s = np.sum(w_s * x_s + b_s)

        y_b = 1/ (1 + np.exp(-h_b))
        y_s = 1/ (1 + np.exp(-h_s))
        
        buy_condition = (y_b>=0.5)
        sell_condition = (y_s>=0.5)

        if buy_condition:
            if not bought:
                btc = usdt / row.close
                usdt = 0
                bought = True
                buy_price += [row.close]
                buy_time += [row.idx]
                n_trades+=1
            if bought:
                pass

        elif sell_condition:
            if bought:
                usdt = btc * row.close
                btc = 0
                sell_price += [row.close]
                sell_time += [row.idx]
                bought = False
                profit += [usdt - usdt_0]
            if not bought:
                pass
    if bought:
        usdt = btc * row.close
        
    pct_better = ((usdt-usdt_0)/usdt_0-buy_and_hold_pct) * 100
    
    plt.figure(figsize=(8,5), dpi=300)
    print("USDT: ", np.round(usdt,2))
    print("%profit ^buy&hold: ", np.round(pct_better,), "%")
    print("n_trades: ", n_trades)
    df.close.plot(linewidth=0.5)
    plt.scatter(buy_time, buy_price, marker="^", c="green", s=100)
    plt.scatter(sell_time, sell_price, marker="v", c="red", s=100)
    
    
```
